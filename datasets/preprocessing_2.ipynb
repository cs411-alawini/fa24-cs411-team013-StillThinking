```python
import pandas as pd
```

```python
job_skills = pd.read_csv('job_skills.csv')
postings = pd.read_csv('postings_processed.csv')
```

```python
df_posting_skills = pd.merge(job_skills, postings, how = 'inner', on = 'job_id')
df_posting_skills[['posting_id', 'skill_abr']]
df_posting_skills[['posting_id', 'skill_abr']].to_csv('posting_skills.csv', index = False)
```

```python
postings['company_id'] = postings['company_id'].astype(int)
postings.to_csv('postings_processed.csv', index = False)
```
